```python
import pandas as pd
```

```python
# Load the data
purchase_orders = pd.read_excel('erp_data/tabPurchase Order.xlsx')
purchase_order_items = pd.read_excel('erp_data/tabPurchase Order Item.xlsx')
purchase_receipts = pd.read_excel('erp_data/tabPurchase Receipt.xlsx')
purchase_receipt_items = pd.read_excel('erp_data/tabPurchase Receipt Item.xlsx')
```

```python


# Join the data
flattable = pd.merge(
    purchase_orders, 
    purchase_order_items, 
    how='left',
    left_on = 'name',
    right_on = 'parent',
    suffixes = ('_po', '_po_items')
    )
```

```python

flattable = pd.merge( 
    flattable, 
    purchase_receipt_items, 
    how='left',
    left_on = 'name_po',
    right_on = 'purchase_order',
    suffixes = ('', '_receipt_items')
    )
```

```python
flattable = pd.merge( 
        flattable, 
        purchase_receipts, 
        how='left',
        left_on = 'parent_receipt_items',
        right_on = 'name',
        suffixes = ('', '_receipt')
        )    
```
